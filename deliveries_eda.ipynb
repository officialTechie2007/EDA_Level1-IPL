## Exploratory Data Analysis for `deliveries.csv`

This notebook follows a three-step EDA workflow using **Pandas** and **Seaborn**:
1. Understand the data
2. Clean the data
3. Analyze relationships in the data

## Initial Setup

Import the required libraries and load the dataset from the project folder.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')

df = pd.read_csv('deliveries.csv')
df
```

## Step 1: Understand the Data

Take an initial look at the dataset, inspect its size, summarize numerical columns, inspect column names, and count unique values across all columns.

```python
# Display the first 5 rows
print('First 5 rows:')
display(df.head())

# Display the last 5 rows
print('Last 5 rows:')
display(df.tail())
```

```python
# Print the shape of the dataset
print(f'Dataset shape: {df.shape}')
```

```python
# Generate statistical summaries for numerical columns
display(df.describe())
```

```python
# List all column names
print('Column names:')
print(df.columns.tolist())
```

```python
# Check the number of unique values in each column
unique_counts = df.nunique().sort_values()
display(unique_counts.to_frame(name='unique_values'))

print('Columns with low cardinality that may be useful for dummy variables:')
potential_dummy_columns = unique_counts[unique_counts <= 10].index.tolist()
print(potential_dummy_columns)
```

## Step 2: Clean the Data

Check for missing values, handle them, remove clearly unnecessary columns, and leave a provision for outlier handling.

```python
# Check for null or missing values
missing_values = df.isnull().sum()
display(missing_values[missing_values > 0].to_frame(name='missing_count'))

# Create a working copy for cleaning
df_clean = df.copy()

# Fill missing values in numeric columns with the mean
numeric_columns = df_clean.select_dtypes(include='number').columns
df_clean[numeric_columns] = df_clean[numeric_columns].fillna(df_clean[numeric_columns].mean())

# Drop rows where important categorical dismissal fields are missing
# This keeps the dataset tidy without forcing artificial labels into these columns.
categorical_columns = ['player_dismissed', 'dismissal_kind', 'fielder']
existing_categorical_columns = [col for col in categorical_columns if col in df_clean.columns]
if existing_categorical_columns:
    df_clean = df_clean.dropna(subset=existing_categorical_columns, how='all')

print(f'Cleaned dataset shape: {df_clean.shape}')
```

```python
# Drop redundant or unnecessary columns that do not add analytical value here
# match_id behaves like an identifier, so we remove it from the main analysis frame.
redundant_columns = ['match_id']
redundant_columns = [col for col in redundant_columns if col in df_clean.columns]
df_clean = df_clean.drop(redundant_columns, axis=1)

print('Remaining columns after dropping redundant fields:')
print(df_clean.columns.tolist())
```

## Step 3: Analyze the Data (Relationship Analysis)

Use Seaborn visualizations to study correlations, pairwise relationships, distributions, and grouped comparisons.

```python
# Correlation matrix heatmap for numerical variables
corr_matrix = df_clean.select_dtypes(include='number').corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Pair plot for a representative sample to keep the notebook responsive
pairplot_columns = ['over', 'ball', 'batsman_runs', 'extra_runs', 'total_runs']
pairplot_sample = df_clean[pairplot_columns].sample(n=min(2000, len(df_clean)), random_state=42)
sns.pairplot(pairplot_sample)
plt.show()
```

```python
# Scatter plot with hue for grouped analysis
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x='over', y='total_runs', hue='inning', alpha=0.6)
plt.title('Scatter Plot of Over vs Total Runs by Inning')
plt.show()
```

```python
# Histogram to visualize the distribution of total runs
plt.figure(figsize=(10, 6))
sns.histplot(data=df_clean, x='total_runs', bins=20, kde=True)
plt.title('Histogram of Total Runs')
plt.show()
```
